### Algoritma Directed Graphs

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pylab

G = nx.DiGraph()

G.add_edges_from([('A','B'),('A','E')], weight=1)
G.add_edges_from([('B','C'),('B','F')], weight=2)
G.add_edges_from([('C','D')], weight=3)
G.add_edges_from([('D','E')], weight=4)
G.add_edges_from([('E','F')], weight=5)
G.add_edges_from([('F','G')], weight=6)
G.add_edges_from([('G','H'),('G','C')], weight=7)
G.add_edges_from([('H','A'),('H','D')], weight=8)


val_map = {'A': 1.61803398875, 'B': 2.71828, 'C': 3.1415926535, 
           'D': 4, 'E': 5, 'F': 6, 'G': 7, 'H': 8}

values = [val_map.get(node, 8) for node in G.nodes()]
edge_labels=dict([((u,v,),d['weight']) for u,v,d in G.edges(data=True)])
edge_colors = ['red' for edge in G.edges()]

pos=nx.spring_layout(G)
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels)
nx.draw(G, pos, node_color =values, node_size=1500, edge_color=edge_colors, edge_cmap=plt.cm.Blues)
node_labels = {node:node for node in G.nodes()}; nx.draw_networkx_labels(G, pos, labels=node_labels)
pylab.show()

print("Total number of nodes: ", int(G.number_of_nodes())) 
print("Total number of edges: ", int(G.number_of_edges())) 
print("List of all nodes: ", list(G.nodes())) 
print("List of all edges: ", list(G.edges(data = True))) 
print("Degree for all nodes: ", dict(G.degree()))
```
